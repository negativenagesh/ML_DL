```python
import pandas as pd
```

```python
data=pd.read_csv('sonar.csv',header=None)
data.tail()
```

```python
from sklearn.utils import shuffle
data=shuffle(data)
```

```python
data[60]=data[60].map({'M':1,'R':0})
data[60].sample(5)
```

```python
x=data.drop(columns=[60])
y=data[60]
```

```python
print(x.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
from sklearn.preprocessing import MinMaxScaler

scale=MinMaxScaler()

x_train_scaled=pd.DataFrame(scale.fit_transform(x_train))
x_test_scaled=pd.DataFrame(scale.fit_transform(x_test))
```

```python
x_train_scaled.head()
```

```python
x_test_scaled.head()
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn=KNeighborsClassifier(n_neighbors=3)
knn.fit(x_train_scaled,y_train)

predicted=knn.predict(x_test_scaled)
print(predicted)
```

```python
from sklearn.metrics  import accuracy_score, classification_report

acc=accuracy_score(y_test,predicted)
print(f"Accuracy score is {acc*100}")
```
